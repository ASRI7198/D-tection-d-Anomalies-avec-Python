## Base de données Mouse

```python
from mouse import *

filepath = "mouse.txt" 

# Load and plot initial data
data, main_points, last_points = load_data(filepath)
plot_data(main_points, last_points, 'Tête de Mickey Mouse avec les vrai outliers')

# Calculate contamination ratio
contamination_ratio = len(last_points) / len(data)
print(f"Contamination utilisé : {contamination_ratio}")

# Isolation Forest
isolation_forest = apply_isolation_forest(data, contamination_ratio)
plot_isolation_forest_scores(data, isolation_forest)

# Local Outlier Factor
lof = apply_lof(data, contamination_ratio)
plot_lof_scores(data, lof)

# numerical results
evaluate_models_with_scores([isolation_forest, lof], data)
```

Selon les résultats visuels et numériques, il semble que l'algorithme LocalOutlierFactor ait correctement détecté les 10 outliers, tandis que l'IsolationForest a mal détecté un point

## Données des cartes de crédits

```python
from credit_fraud_detection import *

file_path = "creditcard.csv"

supervised_experiment(file_path)
```

Les résultats montrent que **Random Forest (RF)** surpasse la **Régression Logistique (LR)** dans toutes les configurations, ce qui en fait le meilleur modèle pour cette tâche. Les courbes de précision-rappel (AUC-PR) indiquent que les approches de stratification (0.87) et de sous-échantillonnage (0.87) offrent les meilleures performances pour RF, tandis que la méthode de rééchantillonnage (oversampling) et le rééquilibrage (balancing) réduisent légèrement la performance (AUC-PR à 0.85). En termes de matrices de confusion, RF avec stratification affiche un excellent équilibre, avec un faible nombre de faux positifs (FP = 5) et de faux négatifs (FN = 17). La sous-échantillonnage améliore légèrement le rappel (FN = 16), tandis que le rééchantillonnage et le rééquilibrage augmentent les erreurs (FP ou FN), compromettant la précision ou le rappel. À l'inverse, LR atteint une AUC-PR de 0.74 avec la stratification et le sous-échantillonnage, mais se dégrade avec le rééchantillonnage et le rééquilibrage (AUC-PR à 0.72). En conclusion, RF avec stratification ou sous-échantillonnage est la méthode la plus efficace, offrant un équilibre optimal entre précision et rappel, tandis que LR est moins adapté pour cette tâche.

```python
from credit_fraud_detection import *

file_path = "creditcard.csv"

unsupervised_experiment(file_path)
```

Les résultats montrent que **Isolation Forest (IF)** surpasse nettement **Local Outlier Factor (LOF)** dans la détection non supervisée des anomalies. Avec des paramètres optimisés (n_estimators=100, max_features=1.0, contamination=0.00172), IF atteint une **AUC-PR de 0.1056**, tandis que LOF, avec ses meilleurs paramètres (n_neighbors=50, contamination=0.005), obtient une **AUC-PR très faible de 0.0045**. La courbe PR confirme cette supériorité, IF présentant une meilleure capacité à équilibrer précision et rappel, tandis que LOF reste inefficace sur ce dataset, probablement en raison de la complexité des données ou de la configuration sous-optimale des voisins locaux. En conclusion, IF est clairement le modèle préféré dans ce cas.
